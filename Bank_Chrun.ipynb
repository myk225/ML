Bank Customer Churn Prediction
Problem Statement

Customer churn, also known as customer attrition, refers to the phenomenon where clients stop doing business with a company or discontinue using its services. The churn rate is the percentage of customers who leave over a specific period. Even a single bad experience can lead a customer to churn, and when a large number of customers leave simultaneously, it can result in both financial losses and reputational damage.

ABC BANK wants to predict customer churn in order to identify potential churners and take proactive measures. The goal is to create a predictive model using machine learning techniques that can accurately forecast whether a customer will exit.

Dataset Description

The dataset contains 10,000 rows and 14 columns, with the target variable being Exited.

Variable	Definition
RowNumber	Unique Row Number
CustomerId	Unique Customer ID
Surname	Surname of a customer
CreditScore	Credit score of the customer
Geography	Geographical location
City_Category	Category of the city (A, B, C)
Gender	Sex of the customer
Age	Age of the customer
Tenure	Number of years as a customer
Balance	Current account balance
NumOfProducts	Number of products the customer holds
HasCrCard	Whether the customer has a credit card (0/1)
IsActiveMember	Whether the customer is active (0/1)
EstimatedSalary	Estimated salary of the customer
Exited	Target variable; 1 if the customer left, 0 otherwise
Workflow

The project follows these steps:

Data Preprocessing

Handling missing values (if any).

Encoding categorical variables (e.g., Gender, Geography, City_Category) using one-hot encoding or label encoding.

Feature scaling (normalization or standardization) for numerical features.

Feature Engineering

Selection of relevant features based on correlation or domain knowledge.

Creation of new features if necessary (e.g., ratio of balance to salary).

Train-Test Split

The dataset is split into 70% training set and 30% test set.

Ensures that the model can generalize to unseen data.

Model Building

Applying machine learning algorithms such as:

Logistic Regression

Decision Trees

Random Forest

Gradient Boosting (optional)

Training the models on the training set.

Model Evaluation

Evaluate the models on both training and test sets.

Metrics used:

Accuracy Score

Confusion Matrix

Classification Report (Precision, Recall, F1-Score)

Compare training and test accuracies to identify potential overfitting.

Prediction

The final model predicts whether a customer will churn (Exited = 1) or stay (Exited = 0).

Key Considerations

Overfitting Check: Comparing training and test accuracy ensures that the model is not overfitting to the training data.

Feature Importance: Tree-based models (Random Forest, Gradient Boosting) can provide insights into which features most influence customer churn.

Business Value: Predicting churn allows the bank to proactively retain customers and reduce revenue loss.

Example Code Snippet
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
